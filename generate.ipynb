```python
from PIL import Image 
from IPython.display import display 
import random
import json
```

```python
# Each image is made up a series of traits
# The weightings for each trait drive the rarity and add up to 100%

background = ["Blue", "White", "Purple", "Red", "Yellow", "Tie Dye 1", "Tie Dye 2"] 
background_weights = [30, 10, 15, 5, 10, 10, 20]

strain = ["Blackberry Kush", "GG4", "Slurricane", "Peanut Butter Breath", "Granddaddy Purple"] 
strain_weights = [30, 40, 15, 10, 5]

eyes = ["Blushing", "Green Eyes", "Wink", "Side Eye", "Wince", "Angry Eyes", "Brown Eyes" , "Suprised", "Tearing Up", "Sad Eyes"] 
eyes_weights = [20, 10, 5, 5, 10, 10, 5, 15, 18, 2]

mouth = ["Happy Mouth", "Super Happy Mouth", "Flirty Mouth", "Gasp", "Sad Mouth", "Nyah"] 
mouth_weights = [30, 40, 10, 10, 5, 5]

legs = ["Standing"] 
legs_weights = [100]

armleft = ["Peace Sign", "Middle Finger", "Arm Down"] 
armleft_weights = [10, 20, 70]

armright = ["Peace Sign", "Middle Finger", "Arm Down"] 
armright_weights = [10, 20, 70]

hats = ["Bow", "None"] 
hats_weights = [10, 90]

shiny = ["Shiny", "Not Shiny"] 
shiny_weights = [1, 99]


# Dictionary variable for each trait. 
# Eech trait corresponds to its file name

background_files = {
    "Blue": "blue",
    "White": "white",
    "Purple": "purple",
    "Red": "red",
    "Yellow": "yellow",
    "Tie Dye 1": "tiedye1",
    "Tie Dye 2": "tiedye2",
}

strain_files = {
    "Blackberry Kush": "leaves1",
    "GG4": "leaves2",
    "Slurricane": "leaves3",
    "Peanut Butter Breath": "leaves4",
    "Granddaddy Purple": "leaves5"   
}

eyes_files = {
    "Blushing": "eyes1",
    "Green Eyes": "eyes2",
    "Wink": "eyes3",
    "Side Eye": "eyes4",
    "Wince": "eyes5",
    "Angry Eyes": "eyes6",
    "Brown Eyes": "eyes7",
    "Suprised": "eyes8",
    "Tearing Up": "eyes9",
    "Sad Eyes": "eyes10",
}

mouth_files = {
    "Happy Mouth": "mouth1",
    "Super Happy Mouth": "mouth2",
    "Flirty Mouth": "mouth3",
    "Gasp": "mouth4",
    "Sad Mouth": "mouth5",
    "Nyah": "mouth6",
}

legs_files = {
    "Standing": "legs1",
}
shiny_files = {
    "Shiny": "shiny1",
    "Not Shiny": "shiny2",
}

armleft_files = {
    "Peace Sign": "armleft1",
    "Middle Finger": "armleft2",
    "Arm Down": "armleft3",
}

armright_files = {
    "Peace Sign": "armright1",
    "Middle Finger": "armright2",
    "Arm Down": "armright3",
}

hats_files = {
    "Bow": "hats1",
    # "Wizard Hat": "hats2",
    "None": "hats3",
}
```

```python
## Generate Traits

TOTAL_IMAGES = 1000 # Number of random unique images we want to generate

all_images = [] 

# A recursive function to generate unique image combinations
def create_new_image():
    
    new_image = {} #

    # For each trait category, select a random trait based on the weightings 
    new_image ["Background"] = random.choices(background, background_weights)[0]
    new_image ["Strain"] = random.choices(strain, strain_weights)[0]
    new_image ["Eyes"] = random.choices(eyes, eyes_weights)[0]
    new_image ["Legs"] = random.choices(legs, legs_weights)[0]
    new_image ["Mouth"] = random.choices(mouth, mouth_weights)[0]
    new_image ["ArmLeft"] = random.choices(armleft, armleft_weights)[0]
    new_image ["ArmRight"] = random.choices(armright, armright_weights)[0]
    new_image ["Hats"] = random.choices(hats, hats_weights)[0]
    new_image ["Shiny"] = random.choices(shiny, shiny_weights)[0]
    
    if new_image in all_images:
        return create_new_image()
    else:
        return new_image
    
    
# Generate the unique combinations based on trait weightings
for i in range(TOTAL_IMAGES): 
    
    new_trait_image = create_new_image()
    
    all_images.append(new_trait_image)
    
```

```python
# Returns true if all images are unique
def all_images_unique(all_images):
    seen = list()
    return not any(i in seen or seen.append(i) for i in all_images)

print("Are all images unique?", all_images_unique(all_images))
```

```python
# Add token Id to each image
i = 0
for item in all_images:
    item["tokenId"] = i
    i = i + 1
```

```python
# Get Trait Counts

background_count = {}
for item in background:
    background_count[item] = 0
    
strain_count = {}
for item in strain:
    strain_count[item] = 0

eyes_count = {}
for item in eyes:
    eyes_count[item] = 0

legs_count = {}
for item in legs:
    legs_count[item] = 0

mouth_count = {}
for item in mouth:
    mouth_count[item] = 0

armleft_count = {}
for item in armleft:
    armleft_count[item] = 0


armright_count = {}
for item in armright:
    armright_count[item] = 0

hats_count = {}
for item in hats:
    hats_count[item] = 0

shiny_count = {}
for item in shiny:
    shiny_count[item] = 0


for image in all_images:
    background_count[image["Background"]] += 1
    strain_count[image["Strain"]] += 1
    eyes_count[image["Eyes"]] += 1
    legs_count[image["Legs"]] += 1
    mouth_count[image["Mouth"]] += 1
    armleft_count[image["ArmLeft"]] += 1
    armright_count[image["ArmRight"]] += 1
    hats_count[image["Hats"]] += 1
    shiny_count[image["Shiny"]] += 1
```

```python
#### Generate Metadata for all Traits 
METADATA_FILE_NAME = './metadata/all-traits.json'; 
with open(METADATA_FILE_NAME, 'w') as outfile:
    json.dump(all_images, outfile, indent=4)
```

```python
  
    
#### Generate Images    
for item in all_images:

    background = Image.open(f'./trait-layers/backgrounds/{background_files[item["Background"]]}.png').convert('RGBA')
    legs = Image.open(f'./trait-layers/backgrounds/{background_files[item["Background"]]}.png').convert('RGBA')
    strain = Image.open(f'./trait-layers/leaves/{strain_files[item["Strain"]]}.png').convert('RGBA')
    eyes = Image.open(f'./trait-layers/eyes/{eyes_files[item["Eyes"]]}.png').convert('RGBA')
    legs = Image.open(f'./trait-layers/legs/{legs_files[item["Legs"]]}.png').convert('RGBA')
    mouth = Image.open(f'./trait-layers/mouth/{mouth_files[item["Mouth"]]}.png').convert('RGBA')
    armright = Image.open(f'./trait-layers/armright/{armright_files[item["ArmRight"]]}.png').convert('RGBA')
    armleft = Image.open(f'./trait-layers/armleft/{armleft_files[item["ArmLeft"]]}.png').convert('RGBA')
    hats = Image.open(f'./trait-layers/hats/{hats_files[item["Hats"]]}.png').convert('RGBA')
    shiny = Image.open(f'./trait-layers/shiny/{shiny_files[item["Shiny"]]}.png').convert('RGBA')

    #Create each composite
    com1 = Image.alpha_composite(background, legs)
    com2 = Image.alpha_composite(com1, armleft)
    com3 = Image.alpha_composite(com2, armright)
    com4 = Image.alpha_composite(com3, strain)
    com5 = Image.alpha_composite(com4, eyes)
    com6 = Image.alpha_composite(com5, mouth)
    com7 = Image.alpha_composite(com6, hats)
    com8 = Image.alpha_composite(com7, shiny)

    #Convert to RGB
    rgb_im = com8.convert('RGB')
    file_name = str(item["tokenId"]) + ".png"
    rgb_im.save("./offering/images/" + file_name)
    
    
    
```

```python
#### Generate Metadata for each Image    

f = open('./metadata/all-traits.json',) 
data = json.load(f)


IMAGES_BASE_URI = "https://gateway.pinata.cloud/ipfs/QmaYNCnNCv2dUEq4hQfVqQETqUgBEB8Yzj9WNZTfuPJdxc/"
PROJECT_NAME = "Lil' Grams"

def getAttribute(key, value):
    return {
        "trait_type": key,
        "value": value
    }
for i in data:
    token_id = i['tokenId']
    token = {
        "image": IMAGES_BASE_URI + str(token_id) + '.png',
        "tokenId": token_id,
        "name": PROJECT_NAME + ' ' + str(token_id),
        "attributes": []
    }
    token["attributes"].append(getAttribute("Background", i["Background"]))
    token["attributes"].append(getAttribute("Strain", i["Strain"]))
    token["attributes"].append(getAttribute("Eyes", i["Eyes"]))
    token["attributes"].append(getAttribute("Legs", i["Legs"]))
    token["attributes"].append(getAttribute("Mouth", i["Mouth"]))
    token["attributes"].append(getAttribute("ArmRight", i["ArmRight"]))
    token["attributes"].append(getAttribute("ArmLeft", i["ArmLeft"]))
    token["attributes"].append(getAttribute("Hats", i["Hats"]))
    token["attributes"].append(getAttribute("Shiny", i["Shiny"]))

    with open('./offering/metadata/' + str(token_id), 'w') as outfile:
        json.dump(token, outfile, indent=4)
f.close()
```
